# Libraries

```python
import bezier
import numpy as np
import seaborn
```

# Smooth Figure

```python
nodes1 = np.asfortranarray([[ 0.2, 1.0, -1.0, 0.1, 0.2],
                            [ 0.5, 0.5, -1.0, 0.4, 0.5],
                            ])
curve1 = bezier.Curve(nodes1, degree=4)
```

```python
seaborn.set()
ax = curve1.plot(num_pts=256)

```
