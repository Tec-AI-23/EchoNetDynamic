```python
import cv2
import numpy as np
import os
import pandas as pd
```

```python
video_info = pd.read_csv('../EchoNet-Dynamic/EchoNet-Dynamic/VolumeTracings.csv')
video_info
```

```python
path = '../EchoNet-Dynamic/EchoNet-Dynamic/Videos/'
files = os.listdir(path)
```

```python
for file in files[: 5]:
    path_file = os.path.join(path, file)
    cap = cv2.VideoCapture(path_file)
    frames = video_info[video_info.FileName == file]['Frame'].unique()
    for frame in frames:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame)
        coor = video_info[(video_info.FileName == file) & (video_info.Frame == frame)]
        puntos1 = [(int(row['X1']), int(row['Y1'])) for index, row in coor.iterrows()]
        puntos2 = [(int(row['X2']), int(row['Y2'])) for index, row in coor.iterrows()]
        ret, img = cap.read()
        if ret:
            for punto in puntos1:
                cv2.circle(img, punto, 1, (0, 255, 0), -1)
            
            for punto in puntos2:
                cv2.circle(img, punto, 1, (0, 255, 0), -1)

            img_rsz = cv2.resize(img, (500, 500))
            cv2.imshow('Image', img_rsz)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
```

```python
for file in files[: 3]:
    path_file = os.path.join(path, file)
    cap = cv2.VideoCapture(path_file)
    frames = video_info[video_info.FileName == file]['Frame'].unique()
    for frame in frames:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame)
        coor = video_info[(video_info.FileName == file) & (video_info.Frame == frame)]
        puntos1 = [(int(row['X1']), int(row['Y1'])) for index, row in coor.iterrows()]
        puntos2 = [(int(row['X2']), int(row['Y2'])) for index, row in coor.iterrows()]
        poly = np.array(np.concatenate((puntos1, puntos2), axis=0).tolist(), np.int32)
        ret, img = cap.read()
        img1 = np.zeros(img.shape[:2],  dtype=np.uint8)
        cv2.fillPoly(img1, [poly], 255)
        img_rsz = cv2.resize(img1, (300, 300))
        cv2.imshow('Image', img_rsz)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```

```python
for file in files[:10]:
    path_file = os.path.join(path, file)
    cap = cv2.VideoCapture(path_file)
    frames = video_info[video_info.FileName == file]['Frame'].unique()
    for frame in frames:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame)
        coor = video_info[(video_info.FileName == file) & (video_info.Frame == frame)]
        puntos1 = [(int(row['X1']), int(row['Y1'])) for index, row in coor.iterrows()]
        puntos2 = [(int(row['X2']), int(row['Y2'])) for index, row in coor.iterrows()]
        puntos2.pop(0)
        puntos2.reverse()
        puntos2.append(puntos1[0])
        poly = np.array(np.concatenate((puntos1, puntos2), axis=0).tolist(), np.int32)
        ret, img = cap.read()
        if ret:
            img1 = np.zeros(img.shape[:2],  dtype=np.uint8)
            cv2.fillPoly(img1, [poly], 255)
            for punto in poly:
                cv2.circle(img, punto, 1, (0, 255, 0), -1)

            name = os.path.join('../EchoNet-Dynamic/EchoNet-Dynamic/masked/', file[0:-4] + '_' + str(frame) + '.jpg')
            cv2.imwrite(name, img1)
            print(name)
            #img_rsz = cv2.resize(img, (500, 500))
            #img1 = cv2.resize(img1, (500, 500))
            #cv2.imshow('Image resized', img_rsz)
            #cv2.imshow('Masked', img1)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
```
