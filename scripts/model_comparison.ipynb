```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mask_model = pd.read_csv("mask_model.csv")
landmark_model = pd.read_csv("landmark_model.csv")
```

```python
# Setup a plot 
plt.figure(figsize=(15, 10))

# Get number of epochs
epochs = range(len(mask_model))

# Plot train loss
plt.subplot(2, 2, 1)
plt.plot(epochs, mask_model["train_loss"], label="mask_model")
plt.plot(epochs, landmark_model["train_loss"], label="landmark_model")
plt.title("Train Loss")
plt.xlabel("Epochs")
plt.legend()

# Plot test loss
plt.subplot(2, 2, 2)
plt.plot(epochs, mask_model["test_loss"], label="mask_model")
plt.plot(epochs, landmark_model["test_loss"], label="landmark_model")
plt.title("Test Loss")
plt.xlabel("Epochs")
plt.legend()

# Plot train accuracy
plt.subplot(2, 2, 3)
plt.plot(epochs, mask_model["train_acc"], label="mask_model")
plt.plot(epochs, landmark_model["train_acc"], label="landmark_model")
plt.title("Train Accuracy")
plt.xlabel("Epochs")
plt.legend()

# Plot test accuracy
plt.subplot(2, 2, 4)
plt.plot(epochs, mask_model["test_acc"], label="mask_model")
plt.plot(epochs, landmark_model["test_acc"], label="landmark_model")
plt.title("Test Accuracy")
plt.xlabel("Epochs")
plt.legend()
```
